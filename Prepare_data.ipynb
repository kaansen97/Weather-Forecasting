##Combined Two Data

```python
df = pd.merge(csv1, csv2, left_on='STA',right_on="WBAN") # Burda TABLE1 + TABLE 2 Yapiyor yani TABLE 2 YI 1.nin altina ekliyor
df=df.dropna(axis=1, how='all')
df.drop('WBAN', axis=1, inplace=True)
df
```

```python
# Correlation
df_corr = df.corr().round(1)  
# Mask to matrix
mask = np.zeros_like(df_corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
# Viz
df_corr_viz = df_corr.mask(mask).dropna(how='all').dropna('columns', how='all')
fig = px.imshow(df_corr_viz, text_auto=True)
fig.show()
```
